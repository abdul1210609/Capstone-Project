### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here

df=df.dropna(subset=['Age' ,'SOPartFreq', 'JobSat'])

age_mapping={ '25-34 years old': 29.5,
             '35-44 years old': 39.5,
             '18-24 years old': 21,
             '45-54 years old': 49.5,
             '55-64 years old': 59.5,
             '65 years or older' : 68,
             'Under 18 years old': 17,
             'Prefer not to say': 0}

df['age_numeric']= df['Age'].map(age_mapping)
df['age_numeric']= pd.to_numeric(df['age_numeric'],errors='coerce')


plt.figure(figsize=(23,10))
sns.scatterplot(data=df, x='age_numeric', y='SOPartFreq', size='JobSat',sizes=(20, 450), alpha=0.7)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here

capvalue = df['ConvertedCompYearly'].quantile(0.97)

plt.figure(figsize=(25,10))
sns.scatterplot(data=df[df['ConvertedCompYearly']< capvalue], x='ConvertedCompYearly', y='JobSat', size='age_numeric', alpha=0.7, hue='JobSat',palette='viridis' )
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction')
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here

df_expanded = df.assign(Language=df['LanguageHaveWorkedWith'].str.split(';')).explode('Language')
df_expanded['Language'] = df_expanded['Language'].str.strip()
df_expanded = df_expanded.dropna(subset=['Language', 'Age'])
top_languages = df_expanded['Language'].value_counts().nlargest(10).index
df_filtered = df_expanded[df_expanded['Language'].isin(top_languages)]
freq= df_filtered['Language'].value_counts()
df_filtered['LanguageCount'] = df_filtered['Language'].map(freq)


plt.figure(figsize=(25,15))
sns.scatterplot(data=df_filtered, x='Language', y='Age', size='LanguageCount', alpha=0.5,sizes=(50, 1500))
plt.title('Language Usage Across Age Groups (Bubble Size by Language Popularity)', fontsize=18)
plt.xlabel('Programming Language', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Popularity Count', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=0.5,labelspacing=0.4,fontsize='medium')
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here

df_expanded = df.assign(Databases=df['DatabaseWantToWorkWith'].str.split(';')).explode('Databases')
df_expanded['Databases'] = df_expanded['Databases'].str.strip()
top_databases = df_expanded['Databases'].value_counts().nlargest(10).index
df_filtered = df_expanded[df_expanded['Databases'].isin(top_databases)] 
df_filtered=df_filtered.dropna(subset=['Databases', 'JobSat', 'ResponseId'])
no_of_respondents = df_filtered.groupby('Databases')['ResponseId'].size()
df_filtered['Number_of_respondents']= df_filtered['Databases'].map(no_of_respondents)



plt.figure(figsize=(25,15))
sns.scatterplot(data=df_filtered, x='Databases', y='JobSat', size='Number_of_respondents', alpha=0.7 ,sizes=(50, 1500))
plt.legend(title='Respondents Count', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=0.5,labelspacing=0.4,fontsize='medium')
plt.title('Relationsship between preferrd databses and Job Satisfaction( Bubbles size by respondents per Database)')
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here

plt.figure(figsize=(25,15))
sns.scatterplot(data=df[df['ConvertedCompYearly'] < capvalue], x='DevType', y='ConvertedCompYearly', size='JobSat', alpha=0.5)
plt.legend(title='Job Sat', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=1,labelspacing=1,fontsize='large')
plt.xlabel('Developer Type')
plt.ylabel('Yearly Compensation')
plt.title('Compensation across different developer roles according to Job Satisfaction')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here

df_expanded = df.assign(Tools=df['NEWCollabToolsHaveWorkedWith'].str.split(';')).explode('Tools')
df_expanded['Tools'] = df_expanded['Tools'].str.strip()
top_tools = df_expanded['Tools'].value_counts().nlargest(10).index
df_filtered = df_expanded[df_expanded['Tools'].isin(top_tools)] 
df_filtered=df_filtered.dropna(subset=['age_numeric', 'Tools'])

freq= df_filtered['Tools'].value_counts()
df_filtered['ToolsCount'] = df_filtered['Tools'].map(freq)


plt.figure(figsize=(25,10))
sns.scatterplot(data=df_filtered, x='Tools', y='age_numeric', size='ToolsCount', alpha=0.7 ,sizes=(50,1000))
plt.legend(title='Tools Usage Count', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=0.5,labelspacing=0.5,fontsize='large')
plt.title('Relationship between top 10 collaboration tools used and age (Bubbles size by of tool usage)')
plt.xlabel('Collaboration Tools Used')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here

df_expanded = df.assign(Frameworks=df['WebframeWantToWorkWith'].str.split(';')).explode('Frameworks')
df_expanded['Frameworks'] = df_expanded['Frameworks'].str.strip()
top_frameworks = df_expanded['Frameworks'].value_counts().nlargest(10).index
df_filtered = df_expanded[df_expanded['Frameworks'].isin(top_frameworks)] 
df_filtered=df_filtered.dropna(subset=['JobSat', 'Frameworks','ResponseId'])

no_of_respondents = df_filtered.groupby('Frameworks')['ResponseId'].size()
df_filtered['Number_of_respondents']= df_filtered['Frameworks'].map(no_of_respondents)

plt.figure(figsize=(25,15))
sns.scatterplot(data=df_filtered, x='Frameworks', y='JobSat', size='Number_of_respondents', alpha=0.5, sizes=(50,1000))
plt.legend(title='Respondents Count', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=0.5,labelspacing=0.5,fontsize='large')
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here

df_expanded = df.assign(Technologies=df['LanguageAdmired'].str.split(';')).explode('Technologies')
df_expanded['Technologies'] = df_expanded['Technologies'].str.strip()
top_technologies = df_expanded['Technologies'].value_counts().nlargest(10).index
top_countries= df_expanded['Country'].value_counts().nlargest(10).index
df_filtered = df_expanded[df_expanded['Technologies'].isin(top_technologies) &df_expanded['Country'].isin(top_countries)]
df_counts = df_filtered.groupby(['Country', 'Technologies']).size().reset_index(name='Admiration_Count')

plt.figure(figsize=(25,10))
sns.scatterplot(data=df_counts, x='Country' ,y='Technologies' , size='Admiration_Count', sizes=(50, 1000), hue='Admiration_Count', palette='magma')
plt.legend(title='Respondents Count', bbox_to_anchor=(1.01, 1), loc='upper left',markerscale=0.5,labelspacing=0.5,fontsize='large')
plt.xticks(rotation=90)
plt.show()
```
