**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
QUERY = '''
SELECT CompTotal FROM main
WHERE CompTotal IS NOT NULL
'''
df=pd.read_sql_query(QUERY,conn)
plt.hist(df, bins=15)
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
QUERY = 'SELECT ResponseId, Age from main'
df=pd.read_sql_query(QUERY, conn)

age_mapping={ '25-34 years old': 29.5,
             '35-44 years old': 39.5,
             '18-24 years old': 21,
             '45-54 years old': 49.5,
             '55-64 years old': 59.5,
             '65 years or older' : 68,
             'Under 18 years old': 17,
             'Prefer not to say': 0}

df['age_numeric']= df['Age'].map(age_mapping)

plt.boxplot(df['age_numeric'])
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here

QUERY = 'SELECT * FROM main'

df=pd.read_sql_query(QUERY, conn)

plt.scatter(df['Age'], df['WorkExp'])
plt.xticks(rotation=90)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here

df_clean = df.dropna(subset=['TimeSearching', 'Frustration', 'Age'])
df_expanded = df_clean.assign(Fru = df['Frustration'].str.split(';')).explode('Fru')
df_expanded['Fru'] = df_expanded['Fru'].str.strip()
topfrustrations = df_expanded['Fru'].value_counts().head(5).reset_index()
df_filtered  = df_expanded[df_expanded['Fru'].isin(topfrustrations['Fru'])]

age_mapping={ '25-34 years old': 29.5,
             '35-44 years old': 39.5,
             '18-24 years old': 21,
             '45-54 years old': 49.5,
             '55-64 years old': 59.5,
             '65 years or older' : 68,
             'Under 18 years old': 17,
             'Prefer not to say': 0}

df_filtered['age_numeric']= df_filtered['Age'].map(age_mapping)

plt.figure(figsize=(15,10))
plt.scatter(df_filtered['TimeSearching'], df_filtered['Fru'], s=df_filtered['age_numeric'],sizes=(50,500), alpha=0.5)
plt.title('Bubble plot between Time searching and frustration according to age')
plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.xticks(rotation=90)
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
top5= df['DatabaseWantToWorkWith'].value_counts().nlargest(5).reset_index()

plt.pie(top5['count'], labels= top5['DatabaseWantToWorkWith'], autopct='%0.2f%%')
plt.title('Top 5 databases respondent want to work with')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here

timesearching_mapping = { '30-60 minutes a day': 45,
                         '15-30 minutes a day': 22.5,
                         '60-120 minutes a day': 90,
                         'Less than 15 minutes a day': 10,
                         'Over 120 minutes a day': 140
                        }

df['TimeSearching_numeric']= df['TimeSearching'].map(timesearching_mapping)

timeanswering_mapping = { '15-30 minutes a day': 22.5,
                         '30-60 minutes a day': 45,
                         'Less than 15 minutes a day': 10,
                         '60-120 minutes a day': 90,
                         'Over 120 minutes a day': 140
                        }
df['TimeAnswering_numeric']=df['TimeAnswering'].map(timeanswering_mapping)

age_mapping={ '25-34 years old': 29.5,
             '35-44 years old': 39.5,
             '18-24 years old': 21,
             '45-54 years old': 49.5,
             '55-64 years old': 59.5,
             '65 years or older' : 68,
             'Under 18 years old': 17,
             'Prefer not to say': 0}

df['age_numeric']= df['Age'].map(age_mapping)


age_group = df.groupby('Age')[['TimeSearching_numeric', 'TimeAnswering_numeric']].agg('median').reset_index()

stacked = age_group[age_group['Age'] == '25-34 years old']
stacked.plot(kind='bar', stacked=True, figsize=(15,10))
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here

df_filtered = df[df['Age'].isin(['45-54 years old', '55-64 years old'])]

mediancomp = df_filtered.groupby('Age')['CompTotal'].median().reset_index()

plt.plot(mediancomp['Age'], mediancomp['CompTotal'])
plt.title('Median Compensation for Ages from 45 to 60')
plt.xlabel('Age Groups')
plt.ylabel('Median Compensation')
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here

df=df.dropna(subset=['MainBranch'])

bar = df['MainBranch'].value_counts()

bar.plot(kind='barh', color='b', figsize=(10,7))
plt.title('Bar Chart for the Main Branch Column')
plt.xlabel('Count')
plt.show()
```
